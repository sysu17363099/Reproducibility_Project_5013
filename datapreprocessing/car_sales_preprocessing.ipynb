```python
df = df.drop(['name','vehicleConfiguration','engineName','engineDisplacement','date','location','link','parse_date', 'description',
               # 'year' this line due to NaN
             ],
        axis=1)
```

```python
df = df.dropna()
df.shape
```

```python
df['brand'].unique().shape, df['bodyType'].unique().shape, df['color'].unique().shape, df['fuelType'].unique().shape, df['transmission'].unique().shape
```

```python
df = df.drop(['brand','color'], axis=1)
```

```python
for c in ['bodyType','fuelType','transmission']:
    df1 = pd.get_dummies(df[c]).astype(int)
    df = pd.concat([df, df1], axis=1)

df = df.drop(['bodyType','fuelType','transmission'], axis=1)
df.shape
```
