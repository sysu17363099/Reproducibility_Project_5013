```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
import pickle
from sklearn import preprocessing
root_path = '/home/xzhang/zxw/good/code_5009'
```

```python
pd.set_option('display.max_columns', 500)  # or 1000
```

```python
file_path = root_path + "/datasets/games.csv"
dataset = pd.read_csv(file_path, delimiter=',')
```

```python
dataset.shape
```

```python
dataset['video_id'].unique().shape
```

```python
dataset = dataset.drop_duplicates(subset='video_id', keep="last")
dataset.reset_index(inplace=True)
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset['categoryId'].unique().shape, dataset['channelTitle'].unique().shape
```

```python
from collections import Counter
tags = []
for ts in dataset['tags'].values:
    tags.extend(ts.split('|'))
c = Counter(tags)
c.most_common(30)
```

```python
ttags = [t for t,cnt in c.most_common(30)[1:]]
```

```python
vs = []
dataset['tags'].values
for ts in dataset['tags'].values:
    vs.append([t in ts for t in ttags])
vs = np.array(vs, dtype=int)
df1 = pd.DataFrame(vs, columns = ttags)
df1
```

```python
df = pd.concat([dataset, df1], axis=1)

df1 = pd.get_dummies(df['categoryId'])
df = pd.concat([df, df1], axis=1)

df = df.drop(['index','video_id','title','publishedAt','categoryId','channelId','channelTitle','trending_date','thumbnail_link','description','tags',
             ],
        axis=1)
df.shape
```

```python
df
```

```python
df['comments_disabled'] = df['comments_disabled'].astype(int)
df['ratings_disabled'] = df['ratings_disabled'].astype(int)
```

```python
df.dtypes.tolist()
```

```python
names = None
items = df

x = items.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
x_scaled = x_scaled / norm(x_scaled, axis=1)[:, None]
items = pd.DataFrame(x_scaled, columns=items.columns)
```

```python
items
```

```python
import os

with open(os.path.expanduser(root_path+'youtube.pkl'), 'wb') as fout:
    pickle.dump([items,names], fout)
```
