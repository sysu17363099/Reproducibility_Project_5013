```python
import pickle
import pandas as pd
import numpy as np
from numpy.linalg import norm
from sklearn import preprocessing
root_path = '/home/xzhang/zxw/good/code_5009'
```

```python
pd.set_option('display.max_columns', 100)  # or 1000
```

```python
file_path = root_path + "/datasets/games.csv"
dataset = pd.read_csv(file_path, delimiter=',')
dataset.head()
```

```python
dataset = dataset.drop(['AppID','Name','Release date','About the game','Full audio languages',
              'Reviews','Header image','Website', 'Support url', 'Support email', 'Metacritic url',
              'Score rank','Notes','Developers','Publishers','Tags', 'Screenshots','Movies',
              # 'Required age','Windows','Mac', 'Linux', 'Supported languages',
             ], axis=1)
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
# drop N/A values 
dataset = dataset.dropna()
# return dataset (row, column)
dataset.shape
```

```python
df = dataset['Supported languages'].str.replace('[','').str.replace(']','').str.replace("'",'').str.replace(' ','').str.get_dummies(sep = ',')
```

```python
# df1.columns.values.tolist()
df = df[['English','French','Japanese','Swedish','Italian','German','Arabic',
           'Finnish','Greek','Hebrew','Persian','Russian','Korean']]
dataset = pd.concat([dataset, df], axis=1)
dataset = dataset.drop(['Supported languages'], axis=1)
dataset.shape
```

```python
dataset['Estimated owners'].unique().shape, dataset['Categories'].unique().shape, dataset['Genres'].unique().shape
```

```python
dataset['Estimated owners'].unique()
```

```python
dataset['Estimated owners'] = dataset['Estimated owners'].str.split(' - ').str[0].astype(int)
```

```python
for column in ['Categories','Genres']:
    temp = dataset[column].str.get_dummies(sep = ',')
    dataset = pd.concat([dataset, temp], axis=1)

dataset = dataset.drop(['Categories','Genres'], axis=1)
dataset.shape
```

```python
for column in ['Windows','Mac', 'Linux']:
    dataset[column] = dataset[column].astype(int)
```

```python
dataset.dtypes.tolist()
```

```python
dataset
```

```python
names = None
items = dataset

x = items.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
x_scaled = x_scaled / norm(x_scaled, axis=1)[:, None]
items = pd.DataFrame(x_scaled, columns=items.columns)
items = items.head(100)
```

```python
items
```

```python
import os

with open(root_path+"/steam_game_100.pkl", 'wb') as fout:
    pickle.dump([items,names], fout)
```
